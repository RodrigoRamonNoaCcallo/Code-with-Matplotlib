## ¿QUÉ ES?


Matplotlib es una librería para la visualización de datos en Python. Fue creada por D. J. Hunter y entre sus características principales estan:
* Su similitud com MATLAB en cuanto a capacidad y sintaxis
* Funciona para programación usual y Programación Orientada a Objetos (POO)
* Sus APIs de POO permite usarla para en desarrollo de aplicaciones

Primero se instala la librería, la fomra más usual es con *pip* o *conda*, comandos ejecutados desde *CMD* o *Anaconda* Prompt respectivamente, luego se importa la librería con *import* (además de matplotlib usaremos la librería numpy)

```python
import matplotlib.pyplot as plt
import numpy as np
```

Existen dos formas de generar escribir el código:
- Usando directamente los comandos o también concido como programación imperativa
- Según la programación orientada a objetos

La documentación de la librería recomienda la segunda opción, que permite un mayor orden cuando la visualización tiene más complejidad

```python
# Programción imperativa
plt.plot([1,2,3,4],[1,2,4,3])
plt.show()
```

```python
# Programación orienta a objetos
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,2,4,3])
plt.show()
```

```python
x = np.linspace(0,2,99)

plt.figure(figsize=(5,2.7), layout="constrained")
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.plot(x,x**3,label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")
plt.legend()

plt.show()
```

```python
x = np.linspace(0,2,99)

fig,ax = plt.subplots(figsize=(5,2.7), layout="constrained")
ax.plot(x,x,label="linear")
ax.plot(x,x**2,label="quadratic")
ax.plot(x,x**3,label="cubic")

ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
plt.legend()
plt.grid()

plt.show()
```

La librería ofrece varias opciones para personalizar el gráfico según diferentes métodos, como se vee en los siguiente ejemplos:

```python
x = np.linspace(0,2*np.pi,49)
y = np.sin(x)

plt.plot(x,y, linewidth=5, ls=(0,(20,5,10,2))) #Se puede usar ls o linestyle
plt.show()
```

```python
x = np.linspace(0,2*np.pi,49)
y = np.sin(x)

fig,ax = plt.subplots()
ax.stem(x,y, linefmt='r:', markerfmt='b*', basefmt='C0-')

ax.spines['left'].set_color('blue')
ax.spines['left'].set_linewidth(3)
ax.spines['left'].set_alpha(0.7)
ax.spines['left'].set_linestyle((0,(2,1)))
ax.spines['left'].set_visible(True)
ax.spines['left'].set_capstyle('butt')

ax.grid(True)
plt.show()
```
